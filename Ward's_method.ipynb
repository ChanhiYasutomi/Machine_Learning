#階層型クラスタリング(Ward's method)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/4-2_skill_level.csv",index_col=0)
```

```python
#階層型クラスタリングを使うにはscipyをimport
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
linkage_result = linkage(df,method="ward",metric="euclidean") #引数->使うデータ入力, 
# method="どういう手法を使うのか", metric="" ->　defaultはユークリッド距離(euclidian) -> 三平方の定理(差の2乗), 困ったらユークリッド距離
# 困ったらward法で良い

linkage_result
```

```python
# ↑のデータを使って可視化
dendrogram(linkage_result,labels=df.index)
plt.show() #それぞれ近似しているグループで分けることができた

# 手法が変わったら結果も変わる
```
