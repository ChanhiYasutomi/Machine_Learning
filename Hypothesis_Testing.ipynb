```python
# 仮説検定 -> 

# データ -> 体力測定の内容,種目によって的する仮設検定手法が変わる
# なにで検定を判断するのか

# 反復横跳び -> 正規分布(平均ぐらいのデータが集まっている)
# 2回目で数値が増えるのか,下がるのか -> 統計的に差があるのか見る場合, T検定
```

```python
import pandas as pd
from scipy import stats #仮説検定する時にimportする

df = pd.read_csv("/content/9-2_Sample.csv",index_col="名前")
```

# 1変数のT検定 -> 標本の平均値が母平均と等しいのかを見る

```python
# 反復横跳び(1回目)が母平均と同等なのかを知りたい
stats.ttest_1samp(df["反復横跳び(1回目)"],popmean=60) #引数にデータを突っ込む,popmean=55世の中の平均

# statistic -> 統計量
# pvalue -> P値(一番大事),0.05より大きいのか(=差はない),小さいのか(差がある)))で見る

# popmean=55の場合 pvalue -> 0.7172290978369188 -> 大きいため差はない,母平均に対して差がないと言える
# popmean=60の場合 pvalue -> 0.0036942099833501985 -> 小さいため差はある,母平均に対して差があると言える

# 両側検定 -> 差があるのかを見る
# 片側検定 -> 大小関係が明白な時に使う

# 片側検定の場合のP値の場合 -> 出たP値を半分にすれば良い
```

# 2変数のT検定 -> 1回目と2回目で差があるのか

```python
# 対応している2変数場合のT検定
stats.ttest_rel(df["反復横跳び(1回目)"],df["反復横跳び(2回目)"]) #変数に比較したいものを2つ入れる

# pvalue=0.006654790139035558 -> 0.05以下のなので差があると言える 
# 片側検定の場合はP値を半分にすれば良い
```

```python
# 男女で差があるのかを見てみる
# ↓男女でできる回数に差があるのかをみる -> 対応していないデータ
stats.ttest_ind(df[df["性別"]=="女性"]["反復横跳び(1回目)"],df[df["性別"]=="男性"]["反復横跳び(1回目)"]) #比較したい2つの変数,分散が同程度の場合⇦のコードを書いて良い,違う場合は違う方が良い
# pvalue=0.15923896868245274 -> 男女で反復横跳びのできる数に差はない
```

# F検定 - 2変数の分散が同程度か,そうでないかを確認する

```python
# 非対応のT検定をする時はF検定で分散が同等なのか,違うのかを調べる
# そのあと分散が同じ場合はequal_var=Falseはいらない,違う場合はequal_var=Falseを使う

# 男女の反復横跳びの分散に差があるのかを調べる
# F検定 -> 2つの分散を同じとしてみなしていいのかがわかる

stats.bartlett(df[df["性別"]=="女性"]["反復横跳び(1回目)"],df[df["性別"]=="男性"]["反復横跳び(1回目)"])
# pvalue=0.2075849593836162 -> 0.05よりも大きいと分散を同じくらいとみなして良い,今回が同程度としてみなして良い
```

```python
# 分散が同程度として見做せない場合 -> 引数に1つ追加する
stats.ttest_ind(df[df["性別"]=="女性"]["反復横跳び(1回目)"],df[df["性別"]=="男性"]["反復横跳び(1回目)"],equal_var=False)
# equal_var=False -> 分散が違うため勘弁して計算してくださいとする
```

```python
# 腕立て伏せはF検定を使っていいいのか怪しい -> 正規分布としてみなせるか怪しいから
# できる人とそうじゃない人がいるので正規分布として見做せるのかは人間側が感覚的に判断しても良い

# T検定 -> 正規分布を仮定した検定
```

# ノンパラメトリック検定 -> 母集団の分布を仮定しない検定

```python
# ノンパラメトリック検定 -> 対応するか,しないかで変わる
# ウイルコクソンの順位符号検定 -> 腕立て伏せで1回目と2回目の回数に差があるのかを見stats.ttest_rel(df["反復横跳び(1回目)"],df["反復横跳び(2回目)"]) #変数に比較したいものを2つ入れる

stats.wilcoxon(df["腕立て伏せ(1回目)"],df["腕立て伏せ(2回目)"]) # ウイルコクソンの順位符号検定
# pvalue=0.015270703918089788 -> 0.05を基準にする, 腕立て伏せの回数に差がある
```

```python
# ウイルコクソンの順位和検(ワンホイットニーのU検定) -> 対応していない場合
stats.mannwhitneyu(df[df["性別"]=="女性"]["腕立て伏せ(1回目)"],df[df["性別"]=="男性"]["腕立て伏せ(1回目)"])
# pvalue=0.015249132623633525 -> 回数に関しては男女でできる回数に差がある
```
