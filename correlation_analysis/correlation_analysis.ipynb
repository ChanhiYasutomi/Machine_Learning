# メモ

パラメトリック統計とノンパラメトリック統計は、統計的検定やモデリングのアプローチにおいて異なる方法論を指します。以下に、それぞれの違いと特徴について説明します。

# パラメトリック統計:

## 違い:

### 仮定する分布:
パラメトリック統計は、データが特定の確率分布（正規分布など）に従っているという仮定を前提とします。

### パラメータ:
統計モデルはパラメータを持ち、これらのパラメータは母集団の性質を表現します。

### データの量:
パラメトリック統計は、比較的多くのデータが利用可能で、これらのデータが仮定した確率分布に合致していることが期待されます。

## 特徴:

### 効率性:
パラメトリック統計は、十分なデータがある場合には非常に効率的であり、パラメータ推定がより精確に行えることがあります。

### 精密さ:
パラメトリックモデルが仮定する分布にデータが合致している場合、パラメトリック統計は非常に精密であり、有力な統計手法として広く使用されています。

# ノンパラメトリック統計:

## 違い:

### 分布の仮定が不要:
ノンパラメトリック統計は、データが特定の確率分布に従っているという仮定を必要としません。したがって、データが非対称であったり、異常値が存在する場合にも適用できます。

### 自由度の少なさ:
パラメトリック統計とは異なり、パラメータの数が少ないか、あるいは全く存在しないモデルも含まれます。

### データの量:
ノンパラメトリック統計は、比較的少ないデータでも適用できる場合があります。

## 特徴:

## 柔軟性:
ノンパラメトリック統計はデータに対して柔軟であり、分布の形状について事前に知識がない場合や、データが非対称である場合に有用です。

##頑健性:
データが外れ値を含んでいたり、分布が非正規分布であったりする場合にも頑健で、ノンパラメトリック手法は頑健性が高いとされています。

選択するべき手法は、データの性質や分布に依存します。データが仮定した分布に従っていると仮定できる場合は、パラメトリック手法が有力ですが、そうでない場合やデータセットが小さい場合は、ノンパラメトリック手法が役立ちます。

# パラメトリック統計の例: t検定

```python
import numpy as np
from scipy.stats import ttest_ind

# 2つの正規分布からサンプルデータを生成
np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=1, scale=1, size=100)

# t検定を実行
t_statistic, p_value = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# P-valueが有意水準（通常は0.05）よりも小さい場合、帰無仮説を棄却できる
if p_value < 0.05:
    print("統計的に有意な差がある")
else:
    print("統計的に有意な差がない")

# この例では、t検定を使用して2つの正規分布から生成されたサンプルデータの平均値に統計的な差があるかどうかを検定しています。
```

# ノンパラメトリック統計の例: ウィルコクソンの符号付き順位検定

```python
from scipy.stats import wilcoxon

# 対応するサンプルデータを生成
before = np.array([25, 30, 35, 40, 45])
after = np.array([30, 35, 35, 40, 50])

# ウィルコクソンの符号付き順位検定を実行
statistic, p_value = wilcoxon(before, after)

print(f"Statistical Value: {statistic}")
print(f"P-value: {p_value}")

# P-valueが有意水準（通常は0.05）よりも小さい場合、帰無仮説を棄却できる
if p_value < 0.05:
    print("統計的に有意な差がある")
else:
    print("統計的に有意な差がない")

# この例では、ウィルコクソンの符号付き順位検定を使用して、前後で対応するサンプルデータの中央値に統計的な差があるかどうかを検定しています。
```

# データが多いと単純に中心極限定理を適応してなんでも正規分布を仮定していいの?

中心極限定理は、サンプルサイズが十分に大きい場合、任意の確率分布から抽出される標本平均は正規分布に収束するという原理です。しかし、この原理はあくまで「大数の法則」の一部であり、いくつかの重要な条件を満たす必要があります。

## 中心極限定理の条件には以下が含まれます：

    無作為標本抽出: サンプルは無作為に、かつ独立に抽出される必要があります。
    同一の確率分布: 各サンプルは同一の確率分布から抽出されている必要があります。
    有限な平均と分散: 抽出される確率分布の平均と分散が有限である必要があります。

さらに、実際のデータが中心極限定理の条件を満たすかどうかは具体的な状況に依存します。確率分布が極端な形状を持つ場合や、外れ値が多い場合など、中心極限定理が適用されないことがあります。

また、データが十分に大きい場合でも、正規分布仮定が厳密に成り立つわけではありません。特に、データが非常に歪んでいたり、外れ値がある場合、正規分布との適合性が検証されるべきです。

統計モデリングや仮説検定を行う際には、中心極限定理の適用性を考慮し、データの性質を検討することが重要です。中心極限定理が適用できない場合や、データが正規分布に適していない場合には、非パラメトリック手法やブートストラップ法などを検討することが一般的です。

#以下が満たされていたら中心極限定理を適応してもいいの?
    無作為標本抽出: サンプルは無作為に、かつ独立に抽出される必要があります。
    同一の確率分布: 各サンプルは同一の確率分布から抽出されている必要があります。
    有限な平均と分散: 抽出される確率分布の平均と分散が有限である必要があります。

条件が満たされていれば、中心極限定理が適用される可能性が高まりますが、依然として注意が必要です。以下のポイントに留意することが重要です：

## サンプルサイズが十分に大きいかどうか:
中心極限定理が効果的に働くためには、十分に大きなサンプルサイズが必要です。通常、サンプルサイズが30以上であれば、中心極限定理が適用されるとされていますが、実際にはより大きなサンプルサイズが好ましいです。

## 非正規分布への適用:
中心極限定理は、無作為標本が同一の確率分布から抽出されている限り、どんな分布でも正規分布に収束するという原理です。しかし、実際のデータが正確にどのような分布に従っているかは確実ではありません。特に極端に歪んだ分布や重い裾を持つ分布の場合、中心極限定理の効果が制限されることがあります。

## 外れ値の影響:
外れ値が存在する場合、中心極限定理の適用性が低下します。外れ値が中心極限定理の仮定を妨げる可能性があるため、注意が必要です。

## 標本が無作為かつ独立であるか:
中心極限定理は、無作為かつ独立な標本が必要です。もし標本が相互に依存している場合、中心極限定理が適用されないことがあります。

要約すると、中心極限定理は便利な統計学の原理ではありますが、注意が必要な条件がいくつかあります。具体的な状況によっては、中心極限定理を用いる前にデータの性質をよく理解し、適切な統計手法を選択することが重要です。

# 無作為標本が同一の確率分布から抽出というのは全ての変数(カラム)が同一の確率分布から抽出されているというのがいいの? 例えば説明変数が10個あれば、それら全てが同一の確率分布から抽出されている必要があるの?

いいえ、全ての変数（カラム）が同一の確率分布から抽出されている必要はありません。中心極限定理が成り立つためには、各変数が同一の確率分布から独立に抽出されているというよりも、各変数の独立性が重要です。

具体的には、無作為標本が同一の確率分布から独立に抽出されている場合、標本の合計や平均などの統計量は中心極限定理に基づいて正規分布に近似されやすくなります。この独立性の仮定が成り立たない場合、中心極限定理の適用性が低下する可能性があります。

例えば、説明変数が10個ある場合、それぞれの説明変数が同じ確率分布から独立に抽出されているか、または相互に独立であることが重要です。ただし、これは厳格な要件ではなく、実際のデータ解析では柔軟性があります。中心極限定理は、標本サイズが大きい場合において、標本統計量が正規分布に近似されるという一般的な原理です。

# データAが発生したことでデータBが発生するようなものはデータAとBの間に関係があるから独立じゃないってことでいいの?

はい、正確です。もしデータAが発生したことが、データBの発生確率に影響を与える場合、それらのデータは互いに従属していると考えられます。このような場合、データAとデータBは独立ではありません。

独立性は統計学や確率論において重要な概念であり、二つの事象（または変数）が互いに独立であれば、一方の事象が発生しても他方の事象には影響を与えないという性質を示します。逆に、二つの事象が独立でない場合、一方の事象が発生することが他方の事象の発生確率に影響を与える可能性があります。

例えば、ある商品の購入がデータAであり、その商品の再購入がデータBである場合、もし再購入確率が初回購入の事実に依存しているならば、データAとデータBは独立ではありません。

# 独立性の検定

```python
# 独立性の検定にはさまざまな手法がありますが、ここではカイ二乗検定 (Chi-squared test) を用いた独立性の検定の例を挙げます。
# カイ二乗検定は、2つのカテゴリカル変数が独立であるかどうかを検定するために使用されます。

# 以下に、Pythonでの具体的な例を示します。例として、性別と好きなスポーツの選択が独立かどうかを検定するケースを考えます。
import pandas as pd
from scipy.stats import chi2_contingency

# ダミーデータ生成
data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Sports': ['Football', 'Football', 'Basketball', 'Basketball', 'Football', 'Basketball']
})

# クロス集計表を作成
observed_table = pd.crosstab(data['Gender'], data['Sports'])

# カイ二乗検定を実行
chi2, p, dof, expected = chi2_contingency(observed_table)

print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies table:")
print(expected)

# P-valueが有意水準（通常は0.05）よりも小さい場合、帰無仮説を棄却できる
if p < 0.05:
    print("性別と好きなスポーツの選択は独立ではない")
else:
    print("性別と好きなスポーツの選択は独立である")

# この例では、カイ二乗検定を用いて性別と好きなスポーツの選択が独立かどうかを検定しています。
# 結果として得られたP-valueが有意水準より小さい場合、帰無仮説（独立性がある）を棄却し、対立仮説（独立性がない）を採用します。
```

# 現実問題として説明変数間がまったくの独立であることはほとんどないように思われる。この場合は説明変数に対して独立性の検定にかけると思うんだけど、これでP-valueが有意水準よりも小さい場合、説明変数間が独立であると仮定してみなしていいの?

説明変数同士がまったくの独立であることは、実際のデータではほとんどありません。多くの場合、説明変数同士は何らかの関連や相互作用を持つことが期待されます。そのため、説明変数同士が完全に独立であるかどうかを検定することは、通常は現実的ではありません。

独立性の検定は、通常、カテゴリカル変数同士の関連性を検証するために使われます。例えば、クロス集計表を用いて2つのカテゴリカル変数の独立性を検証する場合です。しかし、連続変数や混在した型の変数に対して同様の検定を行うことは難しく、適切ではありません。

説明変数同士が相互に独立であるかどうかを確認したい場合、主に以下のアプローチが取られます：

相関係数の確認: 連続変数同士の相関係数を計算し、強い相関がないか確認します。ただし、相関が強い場合でもそれが因果関係を示すわけではないことに注意が必要です。

散布図の可視化: 2つの変数の散布図を描いて、視覚的に相関やパターンを確認します。これにより、変数同士の関連性が明らかになることがあります。

モデルの診断: 変数同士の関連性を評価するためにモデルを構築し、モデルの診断や変数の係数を確認します。

以上の手法を使用して、相関や関連性を評価することが一般的です。ただし、統計的な検定を用いて完全な独立性を検証するのは難しく、適切な手法による探索的なデータ解析が重要です。

# 質的変数には独立性の検定で、量的変数には相関係数を確認すればいいの? それぞれの変数に対して独立性を確認する方法を教えて

はい、おっしゃる通りです。質的変数同士の独立性を確認するためには、主にカイ二乗検定が使用され、量的変数同士の関連性を評価するためには相関係数が一般的に使用されます。

質的変数同士の独立性の確認（カイ二乗検定）:

例えば、2つの質的変数 A と B があり、それらの独立性を検定する場合、カイ二乗検定
を行います。以下はPythonでの例です：

```python
import pandas as pd
from scipy.stats import chi2_contingency

# ダミーデータ生成
data = pd.DataFrame({
    'A': ['Yes', 'No', 'Yes', 'No', 'Yes'],
    'B': ['No', 'Yes', 'Yes', 'No', 'Yes']
})

# クロス集計表を作成
observed_table = pd.crosstab(data['A'], data['B'])

# カイ二乗検定を実行
chi2, p, dof, expected = chi2_contingency(observed_table)

print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies table:")
print(expected)

# P-valueが有意水準（通常は0.05）よりも小さい場合、帰無仮説を棄却できる
if p < 0.05:
    print("変数Aと変数Bは独立ではない")
else:
    print("変数Aと変数Bは独立である")
```

量的変数同士の関連性の確認（相関係数）:

量的変数同士の関連性を確認するためには、相関係数を計算します。

ただし、相関係数は直線的な関係を測定するものであり、非線形の関係には適していません。以下はPythonでの例です：

```python
import pandas as pd

# ダミーデータ生成
data = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 4, 6, 8, 10]
})

# 相関係数を計算
correlation = data['X'].corr(data['Y'])

print(f"Correlation coefficient: {correlation}")

# 通常は相関係数の絶対値が0.7以上で高い相関があるとされるが、具体的な閾値は問題により異なる
if abs(correlation) > 0.7:
    print("変数Xと変数Yは高い相関がある")
else:
    print("変数Xと変数Yは高い相関がない")

# 相関係数の絶対値が0.7以上であれば、高い相関があるとされることがありますが、具体的な閾値は問題により異なります。
```

# 1. t検定やANOVA:

質的変数が2つのカテゴリを持つ場合、t検定を用いて2つのグループの平均値の差を検定することができます。

もし質的変数が3つ以上のカテゴリを持つ場合は、ANOVA (Analysis of Variance) を使用できます。

ANOVAは、複数のカテゴリを持つ質的変数と量的変数の間の差異を評価する統計手法です。複数のカテゴリがある場合、そのカテゴリ間で平均値が統計的に有意に異なるかどうかを検定します。

```python
from scipy.stats import ttest_ind, f_oneway

# ダミーデータ生成
data = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 12, 15, 18, 8, 11]
})

# カテゴリごとにデータを分割
grouped_data = [group['Value'] for name, group in data.groupby('Category')]

# t検定（2つのカテゴリの場合）
t_statistic, p_value = ttest_ind(grouped_data[0], grouped_data[1])

# ANOVA（3つ以上のカテゴリの場合）
f_statistic, p_value_anova = f_oneway(*grouped_data)

print(f"t-statistic: {t_statistic}, p-value: {p_value}")
print(f"F-statistic: {f_statistic}, p-value (ANOVA): {p_value_anova}")
```

# 2. クラメールのV:

クラメールのVは、カテゴリカルな変数同士の関連性を評価する指標で、クロス集計表から計算できます。

値が0に近いほど関連性が低く、1に近いほど高い関連性を示します。

```python
from scipy.stats import chi2_contingency

# ダミーデータ生成
data = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Outcome': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']
})

# クロス集計表を作成
observed_table = pd.crosstab(data['Category'], data['Outcome'])

# カイ二乗検定を実行
chi2, _, _, _ = chi2_contingency(observed_table)

# クラメールのVを計算
num_obs = np.sum(observed_table.values)
v_cramer = np.sqrt(chi2 / (num_obs * (min(observed_table.shape) - 1)))

print(f"Chi-square value: {chi2}")
print(f"Cramer's V: {v_cramer}")
```
