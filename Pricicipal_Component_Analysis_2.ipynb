```python
# 主成分分析 - 古典的手法
# irisデータセットを用いて主成分分析と可視化の方法を説明する．

import pandas as pd
iris = pd.read_csv("http://logopt.com/data/iris.data",
                   names=['sepal length','sepal width','petal length','petal width', 'class']) # 4つの長さを抽出
                   
X = iris[ ['sepal length','sepal width','petal length','petal width'] ]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 2次元に射影,PCA(n_components=次元数指定)
pca.fit(X)
pca.components_ #射影行列
```

```python
Z = pca.transform(X) # 射影したデータ, transform - xを射影するための関数

# 射影した2次元データを元のデータフレームに追加
iris['X']=Z[:,0] #第一成分
iris['Y']=Z[:,1] #第二成分
iris.head()
```

```python
# X,Yを描画
import seaborn as sns
sns.lmplot(x='X',y='Y',hue='class',fit_reg=False,data=iris);

# X,Yを成分だけでだいぶ分けられている
# 主成分分析は古典的な方法でわかりやすいが、あまりいい感じに分けるのが難しい場合が多い
```
